```python
from ProText import gen, low, dig, stopw, lemma, stem, spell, clean_len, WCloud
```

```python
# (low) for converting dataframe to lowercase
# (gen) for Removal of URL, @username, addtional whitespaces, Hashtags, punctuations..etc
# (dig) for removal of numbers
# (stopw) for removal of stopwords
# (clean_len) for removing single or two character word

df['tweets'] = df['tweets'].apply(low).apply(gen).apply(dig).apply(stopw).apply(clean_len)
```
