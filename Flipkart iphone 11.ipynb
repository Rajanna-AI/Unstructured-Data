## Get the iPhone 11 reviews from flipkart


```python
from requests import get
from bs4 import BeautifulSoup

cust_name = []
rev_title = []
rev_rating = []
rev_review = []

# Looping through all the review pages. iphone 11 contains upto 201 pages ofr review
for j in range(1, 201):
    
    link = 'https://www.flipkart.com/apple-iphone-11-white-64-gb/product-reviews/itmfc6a7091eb20b?pid=MOBFWQ6BVWVEH3XE&lid=LSTMOBFWQ6BVWVEH3XESAHPTP&marketplace=FLIPKART&page='
    flipkart = get(link + str(j))
    soup = BeautifulSoup(flipkart.content, 'html.parser')
    
    # Get customer name in every page
    names = soup.find_all('p', class_ = '_2sc7ZR _2V5EHH')
    # Get Title in every page
    title = soup.find_all('p', class_ = '_2-N8zT')
    # Get Rating in every page
    rating = soup.find_all('div', class_ = '_3LWZlK _1BLPMq')
    # Get reviews comments in every page
    review = soup.find_all('div', class_ = 't-ZTKy')    
     
    # Looping through each page element and converting them to list
    for i in range(0, len(rating)):
        cust_name.append(names[i].get_text())
        rev_title.append(title[i].get_text())
        rev_rating.append(rating[i].get_text())
        rev_review.append(review[i].get_text()) 
```

```python
print(len(cust_name),"  ", len(rev_title),"  ", len(rev_rating),"  ", len(rev_review))
```

```python
import pandas as pd

df = pd.DataFrame()

df['Customer Name'] = cust_name
df['Review title'] = rev_title
df['Rating'] = rev_rating
df['Review comment'] = rev_review
df
```

```python
df.to_csv('flipkart_iphone_11.csv')
```
